```python
import pandas as pd

df2=pd.read_csv('D:\Coding\Projects\Hackathon\Data set\Data\Blood_samples_dataset_balanced_2(f).csv')

df2["Disease"]
```

```python

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, drop=None)

disease_encoded = encoder.fit_transform(df2[['Disease']])

disease_columns = [f'Disease_{cat}' for cat in encoder.categories_[0]]

df_encoded_sklearn = pd.DataFrame(
    disease_encoded, 
    columns=disease_columns,
    index=df2.index
)

df_final = pd.concat([df2.drop('Disease', axis=1), df_encoded_sklearn], axis=1)

print(f"New shape: {df_final.shape}")
print(f"\nEncoded columns:")
print(disease_columns)
print(f"\nFirst 5 rows:")
print(df_final.head())


df_final.to_csv('Final model trainer.csv', index=False)
```
